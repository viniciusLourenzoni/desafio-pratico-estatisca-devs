```python
import locale
import pandas as pd
import matplotlib.pyplot as plt
```

```python
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame(dict_faturamento)
```

```python
df_faturamento
```

```python
print(df_faturamento['valor'].mean())
```

```python
media_vendas = df_faturamento['valor'].mean()
```

```python
def formatar_moeda(valor):
    return f'R$ {valor:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')
```

```python
try:
    media_vendas_formatada = locale.currency(media_vendas, grouping=True)
except (locale.Error, NameError):
    media_vendas_formatada = formatar_moeda(media_vendas)
```

```python
# MÃ©dia das vendas
media_vendas_formatada
```
